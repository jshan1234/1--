```python
import random

# ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸
words = ["apple", "banana", "orange", "grape", "lemon"]

def display_word(word, guessed_letters):
    """
    í˜„ì¬ê¹Œì§€ ë§ì¶˜ ê¸€ìë“¤ê³¼ ë°‘ì¤„ë¡œ í‘œì‹œëœ ë‹¨ì–´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
    """
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter + " "
        else:
            display += "_ "
    return display.strip()

def hangman():
    # ëœë¤ìœ¼ë¡œ ë‹¨ì–´ ì„ íƒ
    word = random.choice(words)
    guessed_letters = []
    attempts = 10

    print("í•˜ì´ í–‰ë§¨~!!")
    print("10ë²ˆë§Œ ì‹œë„ í•  ìˆ˜ ìˆì–´ìš”~~")

    while attempts > 0:
        print("\n" + display_word(word, guessed_letters))
        guess = input("ì…ë ¥í•  ë‹¨ì–´: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("ë‹¨ì–´ í•œ ë‹¨ì–´ë§Œ ì³ì£¼ì„¸ìš”~")
            continue

        if guess in guessed_letters:
            print("ë‹¤ë¥¸ ê²ƒì„ ì³ì£¼ì„¸ìš”~")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("ë§ì•˜ìŠµë‹ˆë‹¤~ ê³ ê³ ì”½~~")
        else:
            attempts -= 1
            print(f"í‹€ë ¸ì–´ìš”ã… ã…  {attempts} ë‹¤ì‹œ í•œë²ˆ ì‹œë„~~.")

        if all(letter in guessed_letters for letter in word):
            print("\nì¶•í•˜í•´ìš”! ë§ì¶˜ ë‹¨ì–´ê°€:", word)
            break

    if attempts == 0:
        print("\n ë‹¤ì‹œ ì‹œë„í•´ë³´ì„¸ìš”~", word)

# ê²Œì„ ì‹¤í–‰
hangman()
```
