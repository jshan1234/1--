<span style="font-size:80%">[문제5]<br><br></span>
```Python 2버전대와 3버전대의 차이를 성명해주세요```

<span style="color:red">답변</span>

1.Python 3에서는 print가 함수 형태로 변경되었습니다. Python 2에서는 print "Hello"와 같이 사용했지만, Python 3에서는 print("Hello")와 같이 사용합니다.

2.Python 2에서는 기본적으로 ASCII 문자 집합을 사용하여 문자열을 처리했습니다. 그러나 Python 3에서는 Unicode를 기본 문자열 형식으로 사용합니다. 이것은 문자열 처리 및 유니코드 문자 지원에 대한 개선을 의미합니다.

3.Python 2에서는 next() 함수를 사용하여 다음 값을 가져왔지만, Python 3에서는 next() 함수 대신 --next--() 메서드를 사용합니다.

<span style="font-size:80%">[문제6]<br><br></span>
```Python에서 메모리 관리는 어떻게 이루어지는지 설명해주세요```

<span style="color:red">답변</span>

참조 계수 - 객체를 메모리에 할당할 때 해당 객체에 대한 참조 계수를 추적합니다.객체의 참조 계수가 0이 되면 해당 객체는 더 이상 사용되지 않으며 가비지 컬렉션에 의해 수거됩니다.

가비지 컬렉션가비지 - 참조 계수만으로 관리할 수 없는 순환 참조와 같은 상황을 처리하기 위해 사용됩니다. 순환 참조는 서로가 서로를 참조하는 상황으로, 각각의 객체의 참조 계수가 0이 되지 않습니다.
Python의 가비지 컬렉터는 이러한 순환 참조를 탐지하고 해당 객체들을 메모리에서 제거합니다.
